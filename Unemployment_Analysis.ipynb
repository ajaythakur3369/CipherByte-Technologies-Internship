```python
import calendar
import datetime as dt
```

```python
un2['Date'] = pd.to_datetime(un2['Date'], dayfirst = True)
```

```python
un2['Month'] =  un2['Date'].dt.month
```

```python
un2['Month_int'] = un2['Month'].apply(lambda x : int(x))
```

```python
un2['Month_name'] =  un2['Month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
un2['Region'] = un2['Region'].astype('category')
```

```python
un2.drop(columns = 'Month', inplace = True)
```

```python
un2.head(1)
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
# Assuming 'un2' is your DataFrame
# Step 1: Create a new DataFrame with only numeric columns
numeric_columns = un2.select_dtypes(include = [float, int])

# Step 2: Calculate the correlation matrix using the numeric DataFrame
correlation_matrix = numeric_columns.corr()

# Step 3: Plot a heatmap with the 'coolwarm' colormap
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
sns.boxplot(x = 'Month_int', y = 'Estimated Unemployment Rate', data = un2, width = 0.3)
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Month_int', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Estimated Unemployment Rate', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.show()
```

```python
# Assuming 'un2' is your DataFrame
# Select the columns you want to include in the scatter matrix
columns_to_plot = ['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']

# Create the scatter matrix plot using pandas.plotting.scatter_matrix
scatter_matrix = pd.plotting.scatter_matrix(un2[columns_to_plot], figsize = (25, 20))

# AAdd titles to the subplots (optional)
for ax in scatter_matrix.ravel():
    ax.set_title(ax.get_xlabel() + ' vs. ' + ax.get_ylabel())

# Show the plot
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 6)
plot_ump = un2[['Estimated Unemployment Rate', 'States']]
df_unemp = plot_ump.groupby('States').mean().reset_index()
df_unemp = df_unemp.sort_values('Estimated Unemployment Rate')

fig = px.bar(df_unemp, x = 'States', y = 'Estimated Unemployment Rate', color = 'States',
            title = 'Average unemployment rate in each State', height = 700, template = 'plotly')

fig.show()
```

```python
fig = px.bar(un2, x = 'Region', y = 'Estimated Unemployment Rate', animation_frame = 'Month_name',color = 'States',
            title = 'The unemployment rate across the region from Jan. 2020 to Oct. 2020', height = 900, template = 'plotly')
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000
fig.show()
```
